```python
!ls /content/raw_data/
```

```python
# Run this cell
import os
os.listdir()
```

```python
# Run this cell
raw_files = os.listdir('raw_data')
raw_filepath = ['/content/raw_data/' + file_name for file_name in raw_files]
raw_filepath
```

```python
# Run this cell
df_data = spark.read.csv(raw_filepath, inferSchema=True, header=True)
df_data.show(5)
```

```python
# Run this cell
df_data.printSchema()
```

```python
# Run this cell
df_data.count()
```

```python
# Run this cell
df_data.summary().show()
```

```python
# Run this cell
df_data[['value']].summary().show()
```

```python
# Run this cell
df_data.createOrReplaceTempView('df_data_view')
```

```python
# Run this cell
spark.sql("""
SELECT
  *
FROM
  df_data_view
LIMIT
  5
""").show()
```

```python
spark.sql("""
WITH
value_data AS (
SELECT
  value
FROM
  df_data_view
)
,summary_stats_of_value_data AS (
SELECT
  COUNT(value) AS count
  ,COUNT(DISTINCT value) AS countd
  ,SUM(value) AS sum
  ,AVG(value) AS avg
  ,MIN(value) AS min
  ,PERCENTILE_APPROX(value, 0.25) AS p25
  ,PERCENTILE_APPROX(value, 0.50) AS p50
  ,PERCENTILE_APPROX(value, 0.75) AS p75
  ,MAX(value) AS MAX
  ,COUNT(CASE WHEN value IS NULL THEN 1 ELSE NULL END) AS count_null
  ,COUNT(CASE WHEN value = 0 THEN 1 ELSE NULL END) AS count_zero
FROM
  value_data
)

SELECT
  *
FROM
  summary_stats_of_value_data
""").show()
```
