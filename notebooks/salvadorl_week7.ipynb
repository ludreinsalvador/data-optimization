```python
# Run this cell
spark.sql('''
WITH
year_month_day_split AS (
SELECT
  *
  ,YEAR(datetime) AS year
  ,LPAD(MONTH(datetime), 2, 0) AS month
  ,LPAD(DAY(datetime), 2, 0) AS day
  ,LPAD(WEEKOFYEAR(datetime), 2, 0) AS weeknum
FROM
  df_2178_view
)
,year_month_day_cte AS (
SELECT
  *
  ,CONCAT(year, '-', month) AS year_month
  ,CONCAT(year, '-', month, '-', day) AS year_month_day
  ,CONCAT(year, '-', weeknum) AS year_weeknum
FROM
  year_month_day_split
)
,monthly_average AS (
SELECT
  year_month
  ,AVG(CASE WHEN parameter IN ('so2') THEN value ELSE NULL END) AS so2_mo_avg
  ,AVG(CASE WHEN parameter IN ('co') THEN value ELSE NULL END) AS co_mo_avg
  ,AVG(CASE WHEN parameter IN ('nox') THEN value ELSE NULL END) AS nox_mo_avg
  ,AVG(CASE WHEN parameter IN ('o3') THEN value ELSE NULL END) AS o3_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm10') THEN value ELSE NULL END) AS pm10_mo_avg
  ,AVG(CASE WHEN parameter IN ('no2') THEN value ELSE NULL END) AS no2_mo_avg
  ,AVG(CASE WHEN parameter IN ('no') THEN value ELSE NULL END) AS no_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm25') THEN value ELSE NULL END) AS pm25_mo_avg
FROM
  year_month_day_cte
GROUP BY
  year_month
)
SELECT
  *
FROM
  monthly_average
LIMIT
  5
''').show()
```

```python
# Run this cell
df_mo = spark.sql('''
WITH
year_month_day_split AS (
SELECT
  *
  ,YEAR(datetime) AS year
  ,LPAD(MONTH(datetime), 2, 0) AS month
  ,LPAD(DAY(datetime), 2, 0) AS day
  ,LPAD(WEEKOFYEAR(datetime), 2, 0) AS weeknum
FROM
  df_2178_view
)
,year_month_day_cte AS (
SELECT
  *
  ,CONCAT(year, '-', month) AS year_month
  ,CONCAT(year, '-', month, '-', day) AS year_month_day
  ,CONCAT(year, '-', weeknum) AS year_weeknum
FROM
  year_month_day_split
)
,monthly_average AS (
SELECT
  year_month
  ,AVG(CASE WHEN parameter IN ('so2') THEN value ELSE NULL END) AS so2_mo_avg
  ,AVG(CASE WHEN parameter IN ('co') THEN value ELSE NULL END) AS co_mo_avg
  ,AVG(CASE WHEN parameter IN ('nox') THEN value ELSE NULL END) AS nox_mo_avg
  ,AVG(CASE WHEN parameter IN ('o3') THEN value ELSE NULL END) AS o3_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm10') THEN value ELSE NULL END) AS pm10_mo_avg
  ,AVG(CASE WHEN parameter IN ('no2') THEN value ELSE NULL END) AS no2_mo_avg
  ,AVG(CASE WHEN parameter IN ('no') THEN value ELSE NULL END) AS no_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm25') THEN value ELSE NULL END) AS pm25_mo_avg
FROM
  year_month_day_cte
GROUP BY
  year_month
)
SELECT
  *
FROM
  monthly_average
''')
```

```python
# Run this cell
df_mo_avg = spark.sql('''
WITH
year_month_day_split AS (
SELECT
  *
  ,YEAR(datetime) AS year
  ,LPAD(MONTH(datetime), 2, 0) AS month
  ,LPAD(DAY(datetime), 2, 0) AS day
  ,LPAD(WEEKOFYEAR(datetime), 2, 0) AS weeknum
FROM
  df_2178_view
)
,year_month_day_cte AS (
SELECT
  *
  ,CONCAT(year, '-', month) AS year_month
  ,CONCAT(year, '-', month, '-', day) AS year_month_day
  ,CONCAT(year, '-', weeknum) AS year_weeknum
FROM
  year_month_day_split
)
,monthly_average AS (
SELECT
  year_month
  ,AVG(CASE WHEN parameter IN ('so2') THEN value ELSE NULL END) AS so2_mo_avg
  ,AVG(CASE WHEN parameter IN ('co') THEN value ELSE NULL END) AS co_mo_avg
  ,AVG(CASE WHEN parameter IN ('nox') THEN value ELSE NULL END) AS nox_mo_avg
  ,AVG(CASE WHEN parameter IN ('o3') THEN value ELSE NULL END) AS o3_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm10') THEN value ELSE NULL END) AS pm10_mo_avg
  ,AVG(CASE WHEN parameter IN ('no2') THEN value ELSE NULL END) AS no2_mo_avg
  ,AVG(CASE WHEN parameter IN ('no') THEN value ELSE NULL END) AS no_mo_avg
  ,AVG(CASE WHEN parameter IN ('pm25') THEN value ELSE NULL END) AS pm25_mo_avg
FROM
  year_month_day_cte
GROUP BY
  year_month
)
,monthly_average_with_year_month_day1 AS (
SELECT
   *
   ,CONCAT(year_month, '-01') AS year_month_day1
FROM
  monthly_average
)
,monthly_average_in_2023 AS (
SELECT
  *
FROM
  monthly_average_with_year_month_day1
WHERE
  year_month_day1 BETWEEN '2023-01-01' AND '2023-12-01'
)
SELECT
  *
FROM
  monthly_average_in_2023
''')
```

```python
# Run this cell
df_mo_avg.show()
```

```python
# Run this cell
df_mo_avg.summary().show()
```

```python
# Run this cell
imputer_input = [
    'so2_mo_avg'
    ,'co_mo_avg'
    ,'nox_mo_avg'
    ,'o3_mo_avg'
    ,'pm10_mo_avg'
    ,'no2_mo_avg'
    ,'no_mo_avg'
    ,'pm25_mo_avg'
  ]
imputer_output = [
     'so2_mo_avg_impute'
    ,'co_mo_avg_impute'
    ,'nox_mo_avg_impute'
    ,'o3_mo_avg_impute'
    ,'pm10_mo_avg_impute'
    ,'no2_mo_avg_impute'
    ,'no_mo_avg_impute'
    ,'pm25_mo_avg_impute'
  ]

assembler_input = imputer_output
assembler_output = 'assembler_output'

scaler_input = assembler_output
scaler_output = 'scaler_output'
```

```python
# Run this cell
from pyspark.ml.feature import VectorAssembler, Imputer, StandardScaler
from pyspark.ml.pipeline import Pipeline

imputer = Imputer(inputCols=imputer_input, outputCols=imputer_output)
assembler = VectorAssembler().setInputCols(assembler_input).setOutputCol(assembler_output)
scaler = StandardScaler(inputCol=scaler_input, outputCol=scaler_output)

pipeline = Pipeline(stages=[imputer, assembler, scaler])
```

```python
# Run this cell
pipeline_model = pipeline.fit(df_mo_avg)
```

```python
# Run this cell
df_pipeline = pipeline_model.transform(df_mo_avg)
df_pipeline.show(5)
```

```python
# Run this cell
from pyspark.ml.functions import vector_to_array

df_array = df_pipeline.withColumn('array_scaler_output', vector_to_array("scaler_output"))
df_array.show(5)
```

```python
# Run this cell
df_scale = df_array.withColumn('so2_mo_avg_scale', df_array['array_scaler_output'].getItem(0))\
.withColumn('co_mo_avg_scale', df_array['array_scaler_output'].getItem(1))\
.withColumn('nox_mo_avg_scale', df_array['array_scaler_output'].getItem(2))\
.withColumn('o3_mo_avg_scale', df_array['array_scaler_output'].getItem(3))\
.withColumn('pm10_mo_avg_scale', df_array['array_scaler_output'].getItem(4))\
.withColumn('no2_mo_avg_scale', df_array['array_scaler_output'].getItem(5))\
.withColumn('no_mo_avg_scale', df_array['array_scaler_output'].getItem(6))\
.withColumn('pm25_mo_avg_scale', df_array['array_scaler_output'].getItem(7))

df_scale.show(5)
```

```python
# Run this cell
from pyspark.ml.regression import LinearRegression

assembler = VectorAssembler(inputCols=['co_mo_avg_scale', 'nox_mo_avg_scale', 'o3_mo_avg_scale'], outputCol='features')
df_assembled = assembler.transform(df_scale)

estimator = LinearRegression(featuresCol='features', labelCol='pm10_mo_avg_impute')

model = estimator.fit(df_assembled)

df_prediction = model.transform(df_assembled)
```

```python
# Run this cell
df_prediction.show(5)
```

```python
# Run this cell
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(labelCol='pm10_mo_avg_impute', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(df_prediction)

print('Root Mean Squared Error (RMSE): ', rmse)
print('Coefficients: ', model.coefficients)
print('Intercept: ', model.intercept)
```

## Observation and Analysis

Based on my observations, there were important steps carried out in creating a linear regression model that predicts PM10 concentration from data available on air pollution, including data preparation, feature scaling, model training, and evaluation. First, missing values of air pollutants such as CO, NOx, O3, SO2, NO, PM25, and NO2 were addressed using an imputer and ensured clean data. The features are then aggregated using VectorAssembler, and StandardScaler is used afterwards to scale the feature values so that all of these features, or pollutants, are in the same range. I think that this is a good example of why preprocessing was important to this task because the magnitudes of the different pollutants varied highly and the model needs to be able to handle them properly.

First, the procedure have dealt with missing values for pollutants like CO, NOx, O3 and others, making sure that the data would be neat. Then it applied a technique, called VectorAssembler, to merge all the selected pollutants into one set of features that the model might use. It also applied StandardScaler to normalize the data; the procedure changed all the values to have the same scale so that the model treats every feature equally.

Then it fitted a linear model to predict the levels of PM10, using CO, NOx, and O3 as features. There were some interesting and not so intuitive relationships: an increase in CO lowered the PM10 level, whereas both NOx and O3 had a positive influence, that is to say, if these pollutants increased, then the PM10 would also increase, but for the highest influence on PM10 level, it was O3.

To see how good the model was at predictions, it computed the Root Mean Squared Error, which showed up as 2.911. RMSE is a measure that tells us how far off our predictions were from the true values; on average, the model's predictions were about 2.9 units away from the true PM10 values. It is not perfect, but it does show that the model can give an estimation of levels of PM10.

In general, it appears that O3 is the term that tends to dominate the prediction of the concentrations for PM10, and NOx contributes positively as well. This model would say that reducing the levels of PM10 might best be done by trying to reduce O3. A model like this performs fairly well, but it can be made much better by adding more features, discovering interactions between pollutants, or perhaps even a more complicated model may describe some of these more complicated relationships.
