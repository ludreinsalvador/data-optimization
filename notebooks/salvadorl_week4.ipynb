```python
# Run this cell
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder \
        .master('local[*]') \
        .appName('Basics') \
        .getOrCreate()

print(spark.version)
```

```python
# Run this cell
!aws s3 ls --no-sign-request s3://openaq-data-archive/records/csv.gz/locationid=2178/year=2022/month=05/
```

```python
# Run this cell
!mkdir raw_data
!ls
```

```python
# Run this cell
!aws s3 cp --recursive --no-sign-request s3://openaq-data-archive/records/csv.gz/locationid=2178/year=2022/month=05/ raw_data
```

```python
!ls /content/raw_data/
```

```python
# Run this cell
import os
os.listdir()
```

```python
# Run this cell
raw_files = os.listdir('raw_data')
raw_filepath = ['/content/raw_data/' + file_name for file_name in raw_files]
raw_filepath
```

```python
# Run this cell
df_data = spark.read.csv(raw_filepath, inferSchema=True, header=True)
df_data.show(5)
```

```python
# Run this cell
df_data.printSchema()
```

```python
# Run this cell
df_data.count()
```

```python
# Run this cell
df_data.summary().show()
```

```python
# Run this cell
from pyspark.ml.feature import VectorAssembler, Imputer, StandardScaler
from pyspark.ml.pipeline import Pipeline

impute = Imputer(inputCol='value', outputCol='value_impute')
assembler = VectorAssembler().setInputCols(['value_impute']).setOutputCol('value_impute_vector')
scale = StandardScaler(inputCol='value_impute_vector', outputCol='value_scale')
pipeline = Pipeline(stages=[impute, assembler, scale])
```

```python
# Run this cell
pipeline_model = pipeline.fit(df_data)
```

```python
df_data2 = pipeline_model.transform(df_data)
df_data2.show(2)
```
